```python
# Import pandas and read the csv file
import pandas as pd
data = pd.read_csv("/content/cri t20.csv")
```

```python
# Display the first ten rows
data.head()
```

```python
data.describe()
```

```python
# Bar Plot
import matplotlib.pyplot as plt
x = data.Player
y = data.Inns
plt.bar(x,y,color='g')
plt.title("Bar Plot")
plt.ylabel("Inns")
plt.xticks(rotation='vertical')
plt.xlabel("Player")
plt.show()
```

```python
# Pie Plot
import matplotlib.pyplot as plt
y = data.Runs
plt.pie(y, shadow = True,labels = data.Player)
plt.xlabel("Runs")
plt.title("Pie Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = data.Player
y = data.Runs
plt.scatter(x,y,color='magenta')
plt.title("Scatter plot")
plt.ylabel("Runs")
plt.xticks(rotation='vertical')
plt.xlabel("Player")
```
